```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-7,7,50)
plt.figure(figsize=(16,8))
plt.plot(x,np.cos(x),label='f(x)=cosx',color='red')
plt.plot(x,np.cos(2*x),label='f(x)=cos(2x)',color='green',linestyle='--')
plt.legend(fontsize=14,loc='lower right')
plt.title ('График функции f(x)=cos(Kx)',fontsize=14)
```
